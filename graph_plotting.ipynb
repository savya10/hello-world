```python
from matplotlib.pyplot import *
import numpy as np
import math
```

```python
def variancer(a,b):
    'a = average, b = average of square a. We will use these two to find the variance.'
    variance = b - (a**2)
    return variance

# standard_deviation = (variancer(a,b))**0.5
```

```python
data_8x8 = np.loadtxt('data_6_dt=0.1/8x8.dat')

temp = []
ave_e = []
ave_sq_e = []
ave_m = []
ave_sq_m = []
var_e = []
var_m = []
for i in data8x8:
    temp += [i[0]]
    ave_e += [i[1]/64]
    ave_sq_e += [i[2]/(64**2)]
    ave_m += [i[3]/64]
    ave_sq_m += [i[4]/(64**2)]
    var_e += [variancer((i[1]/64),(i[2]/(64**2)))]
    var_m += [variancer((i[3]/64),(i[4]/(64**2)))]
    
sorted8x8 = [temp,ave_e,ave_sq_e,var_e,ave_m,ave_sq_m,var_m]

std_e_8x8 = np.array(sorted8x8[3])**0.5
std_m_8x8 = np.array(sorted8x8[-1])**0.5

figure(figsize=(30,10))
errorbar(sorted8x8[0], sorted8x8[1], std_e_8x8, ecolor = 'maroon', color = 'blue')
xlabel('Temperature(reduced units)', fontsize = 20)
ylabel('Average Energy per spin (reduced units)', fontsize = 20)
xticks(np.arange(0.5,5.2,0.2), fontsize = 20)
yticks(fontsize = 20)
title('Average energy per spin state for a lattice of size 8x8', fontsize = 20)
show()
```

```python
def C_calcul(a):
    'returns the specific heat capacity at different temperatures for a specific lattice.'
    sq_temp = []
    for i in a[0]:
        sq_temp += [i**2]
    C = []
    for i,sq_t in enumerate(sq_temp):
        C += [(a[3][i])/sq_t]
    return C

heat_cap2x2 = C_calcul(sorted2x2)
heat_cap4x4 = C_calcul(sorted4x4)
heat_cap8x8 = C_calcul(sorted8x8)
heat_cap16x16 = C_calcul(sorted16x16)
heat_cap32x32 = C_calcul(sorted32x32)
```

```python
figure(figsize=(30,10))
plot(sorted32x32[0], heat_cap2x2, label = '2x2')
plot(sorted32x32[0], heat_cap4x4, label = '4x4')
plot(sorted32x32[0], heat_cap8x8, label = '8x8')
plot(sorted32x32[0], heat_cap16x16, label = '16x16')
plot(sorted32x32[0], heat_cap32x32, label = '32x32')
xlabel('Temperature(reduced units)', fontsize = 20)
ylabel('Heat Capacity (reduced units)', fontsize = 20)
xticks(np.arange(0.5,5.2,0.2), fontsize = 20)
yticks(fontsize = 20)
legend(loc = 'best')
title('A plot showing the effect of temperature on the heat capacities of different lattice sizes', fontsize = 20)
show()
```
